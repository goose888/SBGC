```python
# Create new netcdf and store data inside
his = NetCDFFile('CH4_mean_2001_2010.nc', 'w', format='NETCDF4')
his.description = 'ISAM estimation of CH4 emissions for 2001 - 2010.'

nlat = 360 # Size of the matrix ndim*ndim
nlon = 720
xdimension = 0.5
ydimension = 0.5

# Dimensions
his.createDimension('lat', nlat)
his.createDimension('lon', nlon)

# Variables
lat = his.createVariable('lat', 'f4', ('lat',))
lat.long_name = 'latitude'
lat.units = 'degrees_north'
lon = his.createVariable('lon', 'f4', ('lon',))
lon.long_name = 'longitude'
lon.units = 'degrees_east'

ch4_wetland_array = his.createVariable('Emission_wetland', 'f4', ('lat', 'lon',))
ch4_wetland_array.long_name = 'Methane emission from wetland'
ch4_wetland_array.missing_value = -9999.
ch4_wetland_array.units = 'gCH4 m-2 year-1'

ch4_nonwetland_array = his.createVariable('Emission_nonwetland', 'f4', ('lat', 'lon',))
ch4_nonwetland_array.long_name = 'Methane emission from non-wetland soil'
ch4_nonwetland_array.missing_value = -9999.
ch4_nonwetland_array.units = 'gCH4 m-2 year-1'

prod_wetland_array = his.createVariable('Production_wetland', 'f4', ('lat', 'lon',))
prod_wetland_array.long_name = 'Methane production from wetland'
prod_wetland_array.missing_value = -9999.
prod_wetland_array.units = 'gCH4 m-2 year-1'

prod_nonwetland_array = his.createVariable('Production_nonwetland', 'f4', ('lat', 'lon',))
prod_nonwetland_array.long_name = 'Methane production from non-wetland soil'
prod_nonwetland_array.missing_value = -9999.
prod_nonwetland_array.units = 'gCH4 m-2 year-1'

oxid_wetland_array = his.createVariable('Oxidation_wetland', 'f4', ('lat', 'lon',))
oxid_wetland_array.long_name = 'Methane oxidation from wetland'
oxid_wetland_array.missing_value = -9999.
oxid_wetland_array.units = 'gCH4 m-2 year-1'

oxid_nonwetland_array = his.createVariable('Oxidation_nonwetland', 'f4', ('lat', 'lon',))
oxid_nonwetland_array.long_name = 'Methane oxidation from non-wetland soil'
oxid_nonwetland_array.missing_value = -9999.
oxid_nonwetland_array.units = 'gCH4 m-2 year-1'

aere_wetland_array = his.createVariable('Aerenchyma_wetland', 'f4', ('lat', 'lon',))
aere_wetland_array.long_name = 'Methane emission through aerenchyma transport from wetland'
aere_wetland_array.missing_value = -9999.
aere_wetland_array.units = 'gCH4 m-2 year-1'

aere_nonwetland_array = his.createVariable('Aerenchyma_nonwetland', 'f4', ('lat', 'lon',))
aere_nonwetland_array.long_name = 'Methane emission through aerenchyma transport from non-wetland soil'
aere_nonwetland_array.missing_value = -9999.
aere_nonwetland_array.units = 'gCH4 m-2 year-1'

diff_wetland_array = his.createVariable('Diffusion_wetland', 'f4', ('lat', 'lon',))
diff_wetland_array.long_name = 'Methane emission/sink through diffusion from wetland'
diff_wetland_array.missing_value = -9999.
diff_wetland_array.units = 'gCH4 m-2 year-1'

diff_nonwetland_array = his.createVariable('Diffusion_nonwetland', 'f4', ('lat', 'lon',))
diff_nonwetland_array.long_name = 'Methane emission/sink through diffusion from non-wetland soil'
diff_nonwetland_array.missing_value = -9999.
diff_nonwetland_array.units = 'gCH4 m-2 year-1'

ebul_wetland_array = his.createVariable('Ebullition_wetland', 'f4', ('lat', 'lon',))
ebul_wetland_array.long_name = 'Methane emission through ebullition from wetland'
ebul_wetland_array.missing_value = -9999.
ebul_wetland_array.units = 'gCH4 m-2 year-1'

ebul_nonwetland_array = his.createVariable('Ebullition_nonwetland', 'f4', ('lat', 'lon',))
ebul_nonwetland_array.long_name = 'Methane emission through ebullition from non-wetland soil'
ebul_nonwetland_array.missing_value = -9999.
ebul_nonwetland_array.units = 'gCH4 m-2 year-1'

fw_array = his.createVariable('Fractional_water', 'f4', ('lat', 'lon',))
fw_array.long_name = 'Fractional area of surface water'
fw_array.missing_value = -9999.
fw_array.units = '0-1'

# data
lat[:] = np.linspace(-89.75, 89.75, nlat)
lon[:] = np.linspace(0.25, 359.75, nlon)
ch4_wetland_array[:,:] = ncvar_wetland
ch4_nonwetland_array[:,:] = ncvar_nonwtld
prod_wetland_array[:,:] = ncvar_wetland_prod
prod_nonwetland_array[:,:] = ncvar_nonwtld_prod
oxid_wetland_array[:,:] = ncvar_wetland_oxid
oxid_nonwetland_array[:,:] = ncvar_nonwtld_oxid
aere_wetland_array[:,:] = ncvar_wetland_aere
aere_nonwetland_array[:,:] = ncvar_nonwtld_aere
diff_wetland_array[:,:] = ncvar_wetland_diff
diff_nonwetland_array[:,:] = ncvar_nonwtld_diff
ebul_wetland_array[:,:] = ncvar_wetland_ebul
ebul_nonwetland_array[:,:] = ncvar_nonwtld_ebul
fw_array[:,:] = fw

his.close()
```
