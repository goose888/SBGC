```python
from mpl_toolkits.basemap import Basemap, cm, shiftgrid,maskoceans
from netCDF4 import Dataset as NetCDFFile
import numpy as N
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy.ma as ma
import numpy as np
from scipy.interpolate import griddata
from pyresample import geometry,image, kd_tree
import pandas as pd
```

```python
N = 3
res = pd.read_csv('conus_ch4_res.csv')
sourceMeans = res.CH4_source.as_matrix()[[0,2,4]]
oxidMeans = res.CH4_oxidation.as_matrix()[[0,2,4]]
sinkMeans = res.CH4_sink.as_matrix()[[0,2,4]]
netMeans = res.CH4_net.as_matrix()[[0,2,4]]
prodMeans = res.CH4_prod.as_matrix()[[0,2,4]]

sourceStd = res.CH4_source.as_matrix()[[1,3,5]]
oxidStd = res.CH4_oxidation.as_matrix()[[1,3,5]]
sinkStd = res.CH4_sink.as_matrix()[[1,3,5]]
netStd = res.CH4_net.as_matrix()[[1,3,5]]
prodStd = res.CH4_prod.as_matrix()[[1,3,5]]
```

```python
# Read in mean and std for all 4 fluxes and plot the bar chart
# a stacked bar plot with errorbars
# Hist
# Hist_std
# Rcp45
# Rcp45_std
# Rcp85
# Rcp85_std
fig = plt.figure(figsize=(12,12))
plt.rcParams.update({'font.size': 22})
ind = np.arange(N)    # the x locations for the groups
width = 0.1       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind+0.3, sourceMeans, width, color='#d62728', yerr=sourceStd, linewidth=3)
p2 = plt.bar(ind+0.45, sinkMeans, width, yerr=sinkStd) #bottom=menMeans, yerr=womenStd)
p3 = plt.bar(ind+0.6, netMeans, width, yerr=netStd) #bottom=menMeans, yerr=womenStd)
p4 = plt.bar(ind+0.75, oxidMeans, width, yerr=oxidStd) #bottom=menMeans, yerr=womenStd)
p5 = plt.bar(ind+0.9, prodMeans, width, yerr=prodStd) #bottom=menMeans, yerr=womenStd)


#distance = [1,3,7,9]
#energy = [10,20,30,40]
#sigma = [1,3,2,5]

#plt.errorbar(distance, 
#    energy, 
#    sigma, 
#    capsize=5, 
#    elinewidth=2,
#    markeredgewidth=2)

plt.ylabel('$CH_4$ flux (Tg $CH_4$ $year^{-1}$)')
# plt.title('Scores by group and gender')
plt.xticks(ind+0.5, ('2001-2010', '2091-2100, RCP4.5', '2091-2100, RCP8.5'))
#plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0]), ('G', 'U', 'N', 'O', 'P'))


#plt.show()
plt.savefig('./fig6.png')
```
