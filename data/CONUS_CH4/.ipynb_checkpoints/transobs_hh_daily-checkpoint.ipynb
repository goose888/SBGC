```python
import pandas as pd
import numpy as np
import os
os.environ['PROJ_LIB'] = '/data/keeling/a/sshu3/anaconda2/share/proj'
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
from netCDF4 import Dataset as NetCDFFile
from scipy.interpolate import interp1d
import auxiliary_lib as au
import isamcalc_lib as isam
import subprocess
import socplot_lib as socplt
import Fluxtools as flux
```

```python
# First, US-NC4
# US-NC4 does not have the CH4 emission data. All results are not correct here.
nc=NetCDFFile('US-NC4_WT.nc','r')
# FCH4 unit in: nmolCH4 m-2 s-1
ch4_arr = nc.variables['FCH4'][:]
nc.close()
# To gCH4 m-2 s-1 needs to * 16/1000000
# To per day we need to further *3600*24
ch4 = ch4_arr[:,0,0]
ch4[ch4<-100.] = np.float('nan')
ch4_daily = flux.hr2daily(ch4)* 16./1000000.*3600*24
#ch4_daily[ch4_daily<-100.] = np.float('nan')
ccc = pd.DataFrame(data=ch4_daily)
ch4_daily_10d = ccc.rolling(20).mean()

# Open model simulation
fname='/data/jain1/c/sshu3/SBGC/data/CONUS_CH4/site_daily_10_tgas.usnc4'
ch4_sim = pd.read_csv(fname)
ch4_sim_10d = ch4_sim.rolling(20).mean()

# Make plot for US-NC4
plt.plot(ch4_daily_10d[2*365:5*365])
plt.plot(ch4_sim_10d[2*365:5*365])
plt.ylim([-0.8,0.8])
plt.show()
#plt.savefig('US_NC4.png')
```

```python
# NEXT, site US-StJ
nc=NetCDFFile('US-StJ_WT.nc','r')
# FCH4 unit in: nmolCH4 m-2 s-1
ch4_arr = nc.variables['FCH4'][:]
wt_arr = nc.variables['WT'][:]
nc.close()
# To gCH4 m-2 s-1 needs to * 16/1000000
# To per day we need to further *3600*24
ch4 = ch4_arr[:,0,0]
wt = wt_arr[:,0,0]
ch4[ch4<-2000.] = np.float('nan')
ch4[ch4>2000.] = np.float('nan')

ch4_daily = flux.hr2daily(ch4)* 16./1000000000.*3600*24
wt_daily = flux.hr2daily(wt)
#ch4_daily[ch4_daily<-100.] = np.float('nan')
ccc = pd.DataFrame(data=ch4_daily)
ddd = pd.DataFrame(data=wt_daily)
ch4_daily_10d = ccc.rolling(20).mean()
wt_daily_10d = ddd.rolling(20).mean()

# Open model simulation
fname='/data/jain1/c/sshu3/SBGC/data/CONUS_CH4/site_daily_10_tgas.usstj'
ch4_sim = pd.read_csv(fname)
ch4_sim_10d = ch4_sim.rolling(20).mean()
plt.plot(ch4_daily_10d)
plt.plot(ch4_sim_10d)
#plt.ylim([-0.8,0.8])
plt.show()
#plt.savefig('usstj.png')
```
