```python
import isamcalc_lib as isam
import numpy as np
import matplotlib.pyplot as plt
```

```python
dp = isam.get_isam_soildp(10)
depth_mod = isam.get_depth_modifier(2.0, 10, dp[0])
```

```python
t = np.arange(-10,31)
fact_q10_18 = np.zeros(len(t))
fact_q10_20 = np.zeros(len(t))
fact_q10_15 = np.zeros(len(t))
fact_q10_25 = np.zeros(len(t))
for i in np.arange(len(t)):
    fact_q10_18[i] = isam.q10(1.8, t[i])
    fact_q10_15[i] = isam.q10(1.5, t[i])
    fact_q10_20[i] = isam.q10(2.0, t[i])
    fact_q10_25[i] = isam.q10(2.5, t[i])
```

```python
plt.plot(t, fact_q10_15, color="red")
plt.plot(t, fact_q10_18, color="orange")
plt.plot(t, fact_q10_20, color="green")
plt.plot(t, fact_q10_25, color="blue")
#plt.savefig('Q10_curve.png')
```
